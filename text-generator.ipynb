```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
len(tokenizer.word_index)
```

```python
input_seq=[]
for sent in text.split('\n'):
    # print(sent)
    tokenized_sent=tokenizer.texts_to_sequences([sent])[0]
    
    for i in range(1,len(tokenized_sent)):
      n_gram = tokenized_sent[:i+1]  
      input_seq.append(n_gram)
```

```python
input_seq
```

```python
max_len = max([len(x) for x in input_seq])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_seq = pad_sequences(input_seq,maxlen=max_len,padding='pre')
```

```python
padded_input_seq
```

```python
X =  padded_input_seq[:,:-1]
```

```python
y = padded_input_seq[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y,num_classes=1160)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Bidirectional
```

```python
model = Sequential()
model.add(Embedding(1160,100,input_length=62))
model.add(Bidirectional(LSTM(200)))
model.add(Dense(1160,activation='softmax'))
```

```python
max_len
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100)
```
